# **E-Commerce Data**

Baixar:

```python
import pandas as pd
import os

path = "/kaggle/input/ecommerce-data"

# Veja o nome exato do arquivo
print(os.listdir(path))
# Deve mostrar algo como: 'data.csv'
arquivo = "data.csv"  # troque se o nome for diferente

# Leia usando outra codificação
df = pd.read_csv(os.path.join(path, arquivo), encoding="latin1")

# Mostre as primeiras linhas
print(df.head())
```

Analisar:

```python
df.info()       # tipos de cada coluna
df.describe()   # estatísticas numéricas
df.isnull().sum()  # valores nulos por coluna
```

```python
df.head(10)     # primeiras 10 linhas
df.sample(5)    # 5 linhas aleatórias
```

```python
(df['Quantity'] * df['UnitPrice']).sum() #Receita total
```

```python
df[df['Country'] == 'France'] #Filtrar por país
```

```python
df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10) #Top produtos
```

# **Students Performance in Exams**

Baixar:

```python
import kagglehub
import os
import pandas as pd

# 1️⃣ Baixar a versão mais recente do dataset
path = kagglehub.dataset_download("spscientist/students-performance-in-exams")
print("Pasta do dataset:", path)

# 2️⃣ Listar os arquivos disponíveis (é só um CSV)
print(os.listdir(path))
# Deve mostrar algo como: 'StudentsPerformance.csv'

# 3️⃣ Carregar o CSV em um DataFrame
arquivo = "StudentsPerformance.csv"  # confirme o nome exato
df = pd.read_csv(os.path.join(path, arquivo))

# 4️⃣ Conferir as primeiras linhas
print(df.head())
```

Analisar:

```python
df.info()       # tipos de cada coluna
df.describe()   # estatísticas numéricas
df.isnull().sum()  # valores nulos por coluna
```

```python
df.head(10)     # primeiras 10 linhas
df.sample(5)    # 5 linhas aleatórias
```

```python
df[['math score','reading score','writing score']].mean() #Médias gerais
```

```python
df.groupby('gender')[['math score','reading score','writing score']].mean() #Médias por gênero
```

```python
df[['reading score','writing score']].corr() #Relação leitura x escrita
```

# **Walmart Dataset**

Baixar:

```python
import kagglehub
import os
import pandas as pd

# 1️⃣ Baixar a versão mais recente do dataset
path = kagglehub.dataset_download("yasserh/walmart-dataset")
print("Pasta do dataset:", path)

# 2️⃣ Listar os arquivos disponíveis (é só um CSV)
print(os.listdir(path))
# Deve mostrar algo como: ['Walmart.csv']

# 3️⃣ Carregar o CSV em um DataFrame
arquivo = "Walmart.csv"  # confirme o nome exato
df = pd.read_csv(os.path.join(path, arquivo))

# 4️⃣ Conferir as primeiras linhas
print(df.head())
```

Analisar:

```python
df.info()       # tipos de cada coluna
df.describe()   # estatísticas numéricas
df.isnull().sum()  # valores nulos por coluna
```

```python
df.head(10)     # primeiras 10 linhas
df.sample(5)    # 5 linhas aleatórias
```

```python
df['Weekly_Sales'].sum() # Receita total
```

```python
df.groupby('Store')['Weekly_Sales'].mean().sort_values(ascending=False) # Receita média por loja
```
